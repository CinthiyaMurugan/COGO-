# Set Index as a column and adding trade_type categorical column

```python
data['index'] = data.index
```

```python
data['trade_type_mode'] = np.where(data['trade_type'] == 1, 'Domestic', 
         (np.where(data['trade_type'] == 2, 'Import', 'Export')))
```

## Creating new dataframe with index [year and trade_type ] with combained dependent variables 

```python
data_w2l = data[['index','trade_type','sctg2','tons_2018','tons_2019','tons_2020','value_2018','value_2019','value_2020','current_value_2018','current_value_2019','current_value_2020','tmiles_2018','tmiles_2019','tmiles_2020']]
```

```python
data_w2l = pd.wide_to_long(data_w2l, stubnames = ['tons_','value_','current_value_','tmiles_',], i=['index','trade_type','sctg2'], j="year")
data_w2l.head()
```

## Plot shows the max percentage of mode [in, dms and out] transport respect to commodity category 

```python
hue_order= ['Truck','Rail','Water','Air','Multiple modes & mail', 'Pipeline', 'Other and unknown','No domestic mode']
```

```python
inmode_arr = pd.crosstab(index=data.sctg2,columns=data.fr_inmode, values=None).values
```

```python
max_mode_in = []
for val in inmode_arr:
  max_mode_in.append([np.round(val.max()/val.sum()*100,decimals=2),hue_order[val.argmax()]])
```

```python
mode_in_df = pd.DataFrame(max_mode_in, columns=['max_trans_in','max_mode_in'], index=df_commodity.commodity )
mode_in_df.head()
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(x=mode_in_df.max_trans_in, y=mode_in_df.index, hue=mode_in_df.max_mode_in, orient='h' )
```

```python
outmode_arr = pd.crosstab(index=data.sctg2,columns=data.fr_outmode, values=None).values
```

```python
max_mode_out = []
for val in outmode_arr: 
  max_mode_out.append([np.round(val.max()/val.sum()*100,decimals=2),hue_order[val.argmax()]])
```

```python
mode_out_df = pd.DataFrame(max_mode_out,columns=['max_trans_out','max_mode_out'], index=df_commodity.commodity )
mode_out_df.head()
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(x=mode_out_df.max_trans_out, y=mode_out_df.index, hue=mode_out_df.max_mode_out, orient='h' )
```

```python
dmsmode_arr = pd.crosstab(index=data.sctg2,columns=data.dms_mode, values=None).values
```

```python
max_mode_dms = []
for val in dmsmode_arr: 
  max_mode_dms.append([np.round(val.max()/val.sum()*100,decimals=2),hue_order[val.argmax()]])
```

```python
mode_dms_df = pd.DataFrame(max_mode_dms,columns=['max_trans_dms','max_mode_dms'], index=df_commodity.commodity )
mode_dms_df.head()
```

```python
plt.figure(figsize=(20, 12))
sns.barplot(x=mode_dms_df.max_trans_dms, y=mode_dms_df.index, hue=mode_dms_df.max_mode_dms, orient='h' )
```

## Total Miles good are transported in DMS , Import, Export - Goup by year

```python
# Query the database for Quantity greater than 3
data_type_dms = data_w2l.query('trade_type == 1')
data_type_imp = data_w2l.query('trade_type == 2')
data_type_exp = data_w2l.query('trade_type == 3')
```

```python
tmiles_group_sum = data_type_dms.groupby(['sctg2'])['tmiles_'].sum()
plt.figure(figsize=(25, 10))
with sns.axes_style("whitegrid"):
  sns.barplot(y=df_commodity.commodity, x=tmiles_group_sum, orient='h' )
```

```python
tmiles_group_sum = data_type_imp.groupby(['sctg2'])['tmiles_'].sum()
plt.figure(figsize=(25, 10))
with sns.axes_style("whitegrid"):
  sns.barplot(y=df_commodity.commodity, x=tmiles_group_sum, orient='h' )
```

```python
tmiles_group_sum = data_type_exp .groupby(['sctg2'])['tmiles_'].sum()
plt.figure(figsize=(25, 10))
with sns.axes_style("whitegrid"):
  sns.barplot(y=df_commodity.commodity, x=tmiles_group_sum, orient='h' )
```
